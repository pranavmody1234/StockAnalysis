```python
import pandas as pd
```

```python
apple = pd.read_csv('stocks/AAPL.csv')
facebook = pd.read_csv('stocks/FB.csv')
google = pd.read_csv('stocks/GOOGL.csv')
nvidia = pd.read_csv('stocks/NVDA.csv')
tesla = pd.read_csv('stocks/TSLA.csv')
twitter = pd.read_csv('stocks/TWTR.csv')
```

```python
twitter.head()
```

```python
dfs=[apple, facebook, google, nvidia, tesla, twitter]
```

```python
for df in dfs:
    df['MA50']=df.Close.rolling(50).mean()
    df['MA200']=df.Close.rolling(200).mean()
```

```python
apple.head(200)
```

```python
for df in dfs:
    df['Previous Day Close Price'] = df.Close.shift(1)
```

```python
apple.head()
```

```python
for df in dfs:
    df['Change in price'] = df['Close']-df['Previous Day Close Price']
```

```python
apple.head()
```

```python
for df in dfs:
    df['Percent change in price'] = df.Close.pct_change()
```

```python
apple.head()
```

```python
for df in dfs:
    df['Previous day volume'] = df.Volume.shift(1)
```

```python
apple.head()
```

```python
for df in dfs:
    df['Change in Volume'] = df['Volume']- df['Previous day volume']
```

```python
apple.head()
```

```python
for df in dfs:
    df['Percent change in volume'] = df.Volume.pct_change()
```

```python
apple.to_csv('Apple.csv')
facebook.to_csv('Facebook.csv')
google.to_csv('Google.csv')
nvidia.to_csv('Nvidia.csv')
tesla.to_csv('Tesla.csv')
twitter.to_csv('Twitter.csv')
```
